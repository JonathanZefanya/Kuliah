```python
# import modul yang dibutuhkan
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Read data
datas = pd.read_csv('csv_KNN_Liver_CSU_1.csv')
datas.head()
```

```python
# Buat fungsi untuk membagi data yaitu ciri dan label
def devided_data(datas,name_coloums=['ALB','ALP','ALT','AST','BIL','PROT','GGT']):
  ciri = datas[name_coloums]
  label = datas[['Diagnosa']]
  return ciri,label
```

```python
# Memanggil fungsi devided_data
ciri,label = devided_data(datas)
```

```python
ciri
```

```python
label
```

```python
# Split data menjadi data training dan data testing
from sklearn.model_selection import train_test_split
Ciri_train, Ciri_test, Label_train, Label_test = train_test_split(ciri, label, test_size = 0.10, random_state = 0)
```

```python
Ciri_train
```

```python
Ciri_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifer = KNeighborsClassifier(n_neighbors = 1, metric = 'minkowski', p = 2)
classifer.fit(Ciri_train, Label_train)
```

```python
Label_pred = classifer.predict(Ciri_test)
```

```python
Label_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(Label_test, Label_pred)
ac = accuracy_score(Label_test, Label_pred)
```
