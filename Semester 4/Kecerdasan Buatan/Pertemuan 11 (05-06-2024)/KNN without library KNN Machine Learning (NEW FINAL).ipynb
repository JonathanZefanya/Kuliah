```python
import csv
from google.colab import files
print("File Data Training")
uploaded = files.upload()
print("File Data Testing")
uploaded = files.upload()

# Fungsi untuk membaca data dari file CSV
def baca_csv(filename):
    with open(filename, mode='r') as file:
        reader = csv.reader(file)
        header = next(reader)
        data = [row for row in reader]
    return header, data

# Membaca data training dan data testing
train_header, train_data = baca_csv('data_training.csv')
test_header, test_data = baca_csv('data_testing.csv')
```

```python
# Menampilkan 5 data teratas dari data training
print("5 Data Teratas Data Training:")
for kolom in train_data[:5]:
    print(kolom)

# Menampilkan semua data testing
print("\nSemua Data Testing:")
for kolom in test_data:
    print(kolom)
```

```python
import math

# Fungsi untuk menghitung jarak Euclidean
def euclidean_distance(row1, row2):
    jarak = 0.0
    for i in range(len(row1)-1):
                  # Data Testing    # Data Training
        jarak += (float(row1[i]) - float(row2[i]))**2 # Dikali 2 agar hasilnya positif
    return math.sqrt(jarak) # Akar dari hasil jarak penjumlahan agar mendapatkan hasil jarak asli
```

```python
# Fungsi untuk menemukan K tetangga terdekat
def KNN(training_data, test_row, num_neighbors):
    distances = []
    for train_row in training_data:
        dist = euclidean_distance(test_row, train_row)
        distances.append((train_row, dist))
    # Mengurutkan elemen-elemen dalam daftar distances berdasarkan nilai jarak
    distances.sort(key=lambda tup: tup[1])
    neighbors = [distances[i][0] for i in range(num_neighbors)]
    return neighbors
```

```python
from collections import Counter

# Fungsi untuk membuat prediksi
def klasifikasi_prediksi(training_data, test_row, num_neighbors):
    neighbors = KNN(training_data, test_row, num_neighbors)
    output = [row[-1] for row in neighbors]
    prediksi = Counter(output).most_common(1)[0][0]
    return prediksi
```

```python
# Menghitung jarak dan melakukan prediksi untuk setiap data testing
y_true = []
y_pred = []

for test_row in test_data:
    actual_label = test_row[-1]
    y_true.append(actual_label)
    # Menampilkan jarak dari setiap data testing ke semua data training
    print(f"\nJarak dari data testing {actual_label} ke semua data training:")
    distances = []
    for train_row in train_data:
        dist = euclidean_distance(test_row, train_row)
        distances.append((train_row[-1], dist))

    # Urutkan distances berdasarkan jarak (dist)
    sorted_distances = sorted(distances, key=lambda x: x[1])
    for label, dist in sorted_distances:
        print(f"Jarak {actual_label} ke {label} : {dist}")

    # Prediksi menggunakan KNN (K=3, bisa diubah sesuai kebutuhan)
    prediksi = klasifikasi_prediksi(train_data, test_row, 3)
    y_pred.append(prediksi)
    print(f"Hasil prediksi: {prediksi}")
```

```python
# Import modul untuk menghitung cm dan ac
from sklearn.metrics import confusion_matrix, accuracy_score

ac = accuracy_score(y_true, y_pred)
cm = confusion_matrix(y_true, y_pred)
```

```python
print("Confusion Matrix:")
print(cm)
```
